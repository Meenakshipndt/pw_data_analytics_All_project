##### 3D Surface Plot:

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate z values
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a DataFrame
data = {'X': X.flatten(), 'Y': Y.flatten(), 'Z': Z.flatten()}
df = pd.DataFrame(data)

# Create a 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')

# Show the plot
plt.show()
```

### BOKEH ASSIGNMENT

#### 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

##### Sine Wave Plot

```python
from bokeh.plotting import figure, show
import numpy as np

# Generate x and y data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a Bokeh figure
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

# Add a line plot
p.line(x, y, line_width=2)

# Show the plot
show(p)
```

#### 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

##### Scatter Plot

```python
from bokeh.plotting import figure, show
import numpy as np

# Generate random data
x = np.random.rand(100)
y = np.random.rand(100)
sizes = np.random.randint(10, 100, 100)
colors = ["#%06x" % np.random.randint(0, 0xFFFFFF) for _ in range(100)]

# Create a Bokeh figure
p = figure(title="Scatter Plot")

# Add a scatter plot
p.scatter(x, y, size=sizes, fill_color=colors, line_color="black")

# Show the plot
show(p)
```

#### 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears'] counts = [20, 25, 30, 35]

##### Bar Chart

```python
from bokeh.plotting import figure, show
import numpy as np

# Data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a Bokeh figure
p = figure(title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count')

# Add a bar plot
p.vbar(x=fruits, top=counts, width=0.5)

# Show the plot
show(p)
```

#### 4. Create a Bokeh histogram to visualize the distribution of the given data.

##### Histogram

```python
from bokeh.plotting import figure, show
import numpy as np

# Generate random data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency')

# Add a histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="black")

# Show the plot
show(p)
```
